```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
The distplots in plotly python combines various statistical distribution such as histograms, violin plot, box plot etc in the same plot.

To Create a distplots in plotly python, we can either use the px.histogram function from plotly express or use the create_distplot from the plotly figure_factory.


```

```python
For example, the plotly.express function px.histogram can add a subplot with a different statistical representation than the histogram, given by the parameter marginal.
Plotly Express is the easy-to-use, high-level interface to Plotly, which operates on a variety of types of data and produces easy-to-style figures.
```

```python
Create Distplots using plotly Express â€“
To create distplots in plotly express, we can use the px.histogram function.
```

```python
import plotly.express as px
df=px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", marginal="rug")
fig.show()
```

```python
Instead of adding a rug to the plot, we can add a box plot or a violin plot.
```

```python
import plotly.express as px
df=px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", marginal="box")
fig.show()
```

```python
import plotly.express as px
df=px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", marginal="violin")
fig.show()
```

```python
import plotly.express as px
df=px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", marginal="rug",color='sex')
fig.show()
```
